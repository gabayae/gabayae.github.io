## 1. Variables & Data Types

Python is **dynamically typed** -- you do not need to declare a variable's type. The interpreter infers it from the value you assign.

The four fundamental (scalar) types are:

| Type | Example | Description |
|------|---------|-------------|
| `int` | `42` | Whole numbers |
| `float` | `3.14` | Decimal numbers |
| `str` | `"hello"` | Text |
| `bool` | `True` | Logical values (`True` / `False`) |

````python
# Integers
age = 30
num_students = 45

# Floats
temperature = 36.6
pi = 3.14159

# Strings
name = "Alice"
greeting = 'Hello, World!'

# Booleans
is_enrolled = True
has_passed = False

print(age, temperature, name, is_enrolled)
````

````python
# Using type() to inspect types
print(type(age))           # <class 'int'>
print(type(temperature))   # <class 'float'>
print(type(name))          # <class 'str'>
print(type(is_enrolled))   # <class 'bool'>
````

````python
# Type conversion (casting)
x = "42"
print(type(x), x)          # str

x_int = int(x)
print(type(x_int), x_int)  # int

x_float = float(x)
print(type(x_float), x_float)  # float

# Converting number to string
score = 95
score_str = str(score)
print(type(score_str), score_str)

# Boolean conversions -- 0, empty string, None are False; everything else is True
print(bool(0), bool(1), bool(""), bool("hello"), bool(None))
````

## 2. Operators

### 2.1 Arithmetic Operators

| Operator | Meaning | Example |
|----------|---------|---------|
| `+` | Addition | `3 + 2` = 5 |
| `-` | Subtraction | `3 - 2` = 1 |
| `*` | Multiplication | `3 * 2` = 6 |
| `/` | Division (float) | `7 / 2` = 3.5 |
| `//` | Floor division | `7 // 2` = 3 |
| `%` | Modulo | `7 % 2` = 1 |
| `**` | Exponentiation | `3 ** 2` = 9 |

````python
a, b = 17, 5

print(f"{a} + {b}  = {a + b}")
print(f"{a} - {b}  = {a - b}")
print(f"{a} * {b}  = {a * b}")
print(f"{a} / {b}  = {a / b}")
print(f"{a} // {b} = {a // b}")
print(f"{a} % {b}  = {a % b}")
print(f"{a} ** {b} = {a ** b}")
````

### 2.2 Comparison Operators

Comparison operators return `True` or `False`.

````python
x, y = 10, 20

print(f"{x} == {y} : {x == y}")
print(f"{x} != {y} : {x != y}")
print(f"{x} >  {y} : {x > y}")
print(f"{x} <  {y} : {x < y}")
print(f"{x} >= {y} : {x >= y}")
print(f"{x} <= {y} : {x <= y}")
````

### 2.3 Logical Operators

| Operator | Meaning |
|----------|---------|
| `and` | True if **both** operands are True |
| `or` | True if **at least one** operand is True |
| `not` | Negates the operand |

````python
age = 25
has_id = True

# Can this person enter? (must be 18+ AND have ID)
can_enter = age >= 18 and has_id
print(f"Can enter: {can_enter}")

# Is the person a minor OR without ID?
restricted = age < 18 or not has_id
print(f"Restricted: {restricted}")

# Chained comparisons (Pythonic!)
score = 75
print(f"Score between 70 and 80? {70 <= score <= 80}")
````

## 3. String Formatting

Python offers several ways to build strings from variables. We will cover the two most common modern approaches: **f-strings** (Python 3.6+) and the `.format()` method.

````python
name = "Bob"
score = 87.456

# f-strings (recommended)
print(f"Student: {name}, Score: {score}")
print(f"Rounded score: {score:.2f}")
print(f"Name uppercased: {name.upper()}")
print(f"2 + 3 = {2 + 3}")  # expressions inside braces
````

````python
# .format() method
template = "Hello, {}! Your score is {:.1f}%."
print(template.format(name, score))

# Named placeholders
template2 = "Name: {name}, Age: {age}"
print(template2.format(name="Carol", age=29))
````

````python
# Useful string methods
text = "  Hello, Data Science World!  "

print(text.strip())          # Remove leading/trailing whitespace
print(text.strip().lower())  # Lowercase
print(text.strip().upper())  # Uppercase
print(text.strip().replace("World", "Workshop"))
print(text.strip().split(", "))  # Split into a list
print("science" in text.lower())  # Membership test
print(text.strip().startswith("Hello"))
print(text.strip().count("l"))
````

## 4. Control Flow

Control flow lets your program make decisions. Python uses `if`, `elif`, and `else` blocks, delimited by **indentation** (4 spaces by convention).

````python
temperature = 32

if temperature >= 35:
    print("It's very hot! Stay hydrated.")
elif temperature >= 25:
    print("It's warm and pleasant.")
elif temperature >= 15:
    print("It's cool. Grab a jacket.")
else:
    print("It's cold! Bundle up.")
````

````python
# Practical example: BMI classifier
weight_kg = 70
height_m = 1.75

bmi = weight_kg / (height_m ** 2)

if bmi < 18.5:
    category = "Underweight"
elif bmi < 25:
    category = "Normal weight"
elif bmi < 30:
    category = "Overweight"
else:
    category = "Obese"

print(f"BMI: {bmi:.1f} -> {category}")
````

````python
# Ternary (conditional) expression
age = 20
status = "adult" if age >= 18 else "minor"
print(f"Age {age}: {status}")

# Nested conditions
score = 85
attendance = 90

if score >= 60:
    if attendance >= 75:
        print("Passed with good attendance.")
    else:
        print("Passed but attendance is low.")
else:
    print("Failed.")
````

## 5. Loops

### 5.1 `for` loops

A `for` loop iterates over the elements of any **iterable** (list, string, range, etc.).

````python
# Iterating over a list
fruits = ["apple", "banana", "cherry", "date"]

for fruit in fruits:
    print(f"I like {fruit}s!")
````

````python
# Using range()
print("range(5):       ", list(range(5)))        # 0, 1, 2, 3, 4
print("range(2, 8):    ", list(range(2, 8)))     # 2, 3, 4, 5, 6, 7
print("range(0, 10, 2):", list(range(0, 10, 2))) # 0, 2, 4, 6, 8

# Summing numbers 1 to 100
total = 0
for i in range(1, 101):
    total += i
print(f"\nSum of 1 to 100 = {total}")
````

````python
# enumerate() gives you the index AND the value
students = ["Alice", "Bob", "Carol", "David"]

for idx, student in enumerate(students):
    print(f"  {idx + 1}. {student}")

print()

# enumerate with a custom start
for rank, student in enumerate(students, start=1):
    print(f"  Rank {rank}: {student}")
````

### 5.2 `while` loops

A `while` loop runs as long as its condition is `True`. Be careful to avoid infinite loops!

````python
# Countdown
count = 5
while count > 0:
    print(count, end=" ")
    count -= 1
print("Go!")
````

````python
# break and continue
# Find the first number divisible by 7 in a range
for n in range(1, 100):
    if n % 7 == 0:
        print(f"First multiple of 7: {n}")
        break

# Skip even numbers, print only odd
print("Odd numbers under 10: ", end="")
for n in range(1, 10):
    if n % 2 == 0:
        continue
    print(n, end=" ")
print()
````

## 6. List Comprehensions

List comprehensions are a concise, Pythonic way to create lists. The general syntax is:

```python
[expression for item in iterable if condition]
```

````python
# Basic comprehension: squares of 0-9
squares = [x ** 2 for x in range(10)]
print("Squares:", squares)

# Equivalent traditional loop
squares_loop = []
for x in range(10):
    squares_loop.append(x ** 2)
print("Same:   ", squares_loop)
````

````python
# Conditional comprehension: even numbers only
evens = [x for x in range(20) if x % 2 == 0]
print("Evens:", evens)

# Conditional expression inside comprehension
labels = ["even" if x % 2 == 0 else "odd" for x in range(8)]
print("Labels:", labels)

# String processing with comprehension
words = ["  hello ", " WORLD", "Python  "]
cleaned = [w.strip().lower() for w in words]
print("Cleaned:", cleaned)
````

````python
# Nested comprehension: flattening a matrix
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

flat = [val for row in matrix for val in row]
print("Flattened:", flat)

# Dictionary comprehension (preview)
word = "mississippi"
char_counts = {ch: word.count(ch) for ch in set(word)}
print("Character counts:", char_counts)
````

## 7. Functions

Functions let you encapsulate reusable blocks of logic. In Python, you define them with the `def` keyword.

````python
# Basic function
def greet(name):
    """Return a greeting string."""
    return f"Hello, {name}! Welcome to the workshop."

message = greet("Alice")
print(message)
print(greet("Bob"))
````

````python
# Default arguments
def power(base, exponent=2):
    """Raise base to the given exponent (default: squared)."""
    return base ** exponent

print(power(5))       # 25  (uses default exponent=2)
print(power(5, 3))    # 125
print(power(2, 10))   # 1024

# Multiple return values
def min_max(numbers):
    """Return the minimum and maximum of a list."""
    return min(numbers), max(numbers)

lo, hi = min_max([4, 7, 1, 9, 3])
print(f"Min: {lo}, Max: {hi}")
````

````python
# *args and **kwargs
def summarize(*args, **kwargs):
    """
    Demonstrate variable-length arguments.
    *args   -> tuple of positional arguments
    **kwargs -> dict of keyword arguments
    """
    print(f"Positional args: {args}")
    print(f"Keyword args:    {kwargs}")
    
    if args:
        print(f"Sum of positional args: {sum(args)}")

summarize(1, 2, 3, name="Alice", course="Data Science")

print()

# Practical use: flexible averaging function
def average(*values):
    """Compute the average of any number of values."""
    if not values:
        return 0.0
    return sum(values) / len(values)

print(f"Average: {average(85, 90, 78, 92, 88):.1f}")
````

## 9. Lab 1: Student Grade Analyzer

Now let us put everything together! You will build a small program that:

1. Takes a dictionary of student names and their scores  
2. Computes the **mean**, **median**, **minimum**, and **maximum**  
3. Assigns a **letter grade** to each student  
4. Prints a formatted report  

We will first build it step-by-step, then you will enhance it on your own.

````python
# Step 1: Define the data
student_scores = {
    "Alice":   92,
    "Bob":     78,
    "Carol":   85,
    "David":   63,
    "Eve":     95,
    "Frank":   71,
    "Grace":   88,
    "Hector":  54,
    "Ivy":     79,
    "Jack":    90,
}

print(f"Number of students: {len(student_scores)}")
print(f"Students: {list(student_scores.keys())}")
````

````python
# Step 2: Compute statistics
scores = list(student_scores.values())

def compute_mean(values):
    """Compute the arithmetic mean."""
    return sum(values) / len(values)

def compute_median(values):
    """Compute the median."""
    sorted_vals = sorted(values)
    n = len(sorted_vals)
    mid = n // 2
    if n % 2 == 0:
        return (sorted_vals[mid - 1] + sorted_vals[mid]) / 2
    else:
        return sorted_vals[mid]

mean_score = compute_mean(scores)
median_score = compute_median(scores)
min_score = min(scores)
max_score = max(scores)

print(f"Mean:   {mean_score:.2f}")
print(f"Median: {median_score:.2f}")
print(f"Min:    {min_score}")
print(f"Max:    {max_score}")
````

````python
# Step 3: Assign letter grades
def assign_grade(score):
    """Assign a letter grade based on numeric score."""
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

# Build a results list
results = []
for name, score in student_scores.items():
    grade = assign_grade(score)
    results.append((name, score, grade))

# Sort by score descending
results.sort(key=lambda x: x[1], reverse=True)

# Preview
for name, score, grade in results[:3]:
    print(f"{name}: {score} ({grade})")
````

````python
# Step 4: Print the formatted report
print("=" * 50)
print("         STUDENT GRADE REPORT")
print("=" * 50)
print(f"{'Rank':<6}{'Name':<12}{'Score':>6}{'Grade':>8}")
print("-" * 50)

for rank, (name, score, grade) in enumerate(results, start=1):
    print(f"{rank:<6}{name:<12}{score:>6}{grade:>8}")

print("-" * 50)
print(f"{'Mean:':<18}{mean_score:>6.2f}")
print(f"{'Median:':<18}{median_score:>6.2f}")
print(f"{'Min:':<18}{min_score:>6}")
print(f"{'Max:':<18}{max_score:>6}")
print("=" * 50)

# Grade distribution
grade_counts = {}
for _, _, grade in results:
    grade_counts[grade] = grade_counts.get(grade, 0) + 1

print("\nGrade Distribution:")
for grade in ["A", "B", "C", "D", "F"]:
    count = grade_counts.get(grade, 0)
    bar = "#" * (count * 3)
    print(f"  {grade}: {bar} ({count})")
````
