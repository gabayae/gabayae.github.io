## 6.1 Introduction to Plotting in Julia

Julia has several plotting ecosystems:
- **Plots.jl** -- high-level, multiple backends (GR, Plotly, PyPlot)
- **Makie.jl** -- high-performance, GPU-accelerated, publication-quality
- **StatsPlots.jl** -- statistical plotting recipes
- **UnicodePlots.jl** -- terminal-based plots

## 6.2 Getting Started with Plots.jl

```julia
using Pkg
Pkg.add(["Plots", "StatsPlots"])
```

```julia
using Plots

# Simple line plot
x = 0:0.1:2π
y = sin.(x)
plot(x, y, label="sin(x)", title="Sine Wave",
     xlabel="x", ylabel="y", linewidth=2)
```

```julia
# Multiple series
x = 0:0.1:2π
plot(x, sin.(x), label="sin(x)")
plot!(x, cos.(x), label="cos(x)")  # plot! adds to existing figure
plot!(title="Trigonometric Functions", legend=:bottomleft)
```

```julia
# Scatter plot
n = 50
xs = randn(n)
ys = 2 .* xs .+ randn(n) .* 0.5
scatter(xs, ys, label="data", xlabel="x", ylabel="y",
        title="Scatter Plot", markersize=4, alpha=0.7)
```

## 6.3 Plot Types

```julia
# Bar plot
languages = ["Julia", "Python", "R", "Matlab"]
speed = [1.0, 50.0, 200.0, 30.0]
bar(languages, speed, label="Relative Time",
    title="Benchmark Comparison", ylabel="Time (normalized)",
    color=[:blue, :green, :red, :orange])
```

```julia
# Histogram
data = randn(1000)
histogram(data, bins=30, label="N(0,1)", alpha=0.7,
          title="Histogram of Standard Normal",
          xlabel="Value", ylabel="Frequency")
```

```julia
# Heatmap
A = rand(10, 10)
heatmap(A, title="Random Heatmap", color=:viridis,
        xlabel="Column", ylabel="Row")
```

```julia
# Contour plot
f(x, y) = sin(x) * cos(y)
xs = range(-π, π, length=100)
ys = range(-π, π, length=100)
contour(xs, ys, f, fill=true, title="sin(x)cos(y)",
        color=:turbo, levels=20)
```

## 6.4 Subplots and Layouts

```julia
# Subplots with layout
x = 0:0.1:2π
p1 = plot(x, sin.(x), title="sin(x)")
p2 = plot(x, cos.(x), title="cos(x)")
p3 = plot(x, tan.(x), title="tan(x)", ylims=(-5, 5))
p4 = plot(x, x.^2, title="x²")

plot(p1, p2, p3, p4, layout=(2, 2), size=(800, 600))
```

## 6.5 Customization

```julia
# Customizing appearance
using Plots
x = 0:0.01:4π

plot(x, sin.(x),
    title="Publication-Quality Plot",
    xlabel="x",
    ylabel="f(x)",
    label="sin(x)",
    linewidth=2,
    linestyle=:solid,
    color=:blue,
    grid=true,
    gridstyle=:dash,
    gridalpha=0.3,
    legend=:topright,
    size=(700, 400),
    dpi=300
)
```

## 6.6 Statistical Plots with StatsPlots

```julia
using StatsPlots

# Box plot
data = [randn(50) .+ i for i in 1:4]
boxplot(["A" "B" "C" "D"], hcat(data...),
        title="Box Plots", ylabel="Value", legend=false)
```

```julia
# Violin plot
data = [randn(50) .+ i for i in 1:4]
violin(["A" "B" "C" "D"], hcat(data...),
       title="Violin Plots", ylabel="Value", legend=false)
```

## 6.7 Introduction to Makie.jl

Makie is Julia's high-performance plotting ecosystem.
- `CairoMakie` for publication-quality static plots
- `GLMakie` for interactive/3D
- `WGLMakie` for web

```julia
# using Pkg
# Pkg.add("CairoMakie")

using CairoMakie

# Basic line plot
fig = Figure(size=(800, 500))
ax = Axis(fig[1, 1],
    title="Makie Line Plot",
    xlabel="x",
    ylabel="y"
)
x = 0:0.01:2π
lines!(ax, x, sin.(x), label="sin(x)", linewidth=2)
lines!(ax, x, cos.(x), label="cos(x)", linewidth=2)
axislegend(ax)
fig
```

```julia
# Makie heatmap
using CairoMakie

fig = Figure(size=(600, 500))
ax = Axis(fig[1, 1], title="Heatmap")
data = [sin(x) * cos(y) for x in range(-π, π, 50), y in range(-π, π, 50)]
heatmap!(ax, data, colormap=:viridis)
Colorbar(fig[1, 2])
fig
```
