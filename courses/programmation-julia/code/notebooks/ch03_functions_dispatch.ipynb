## 3.1 Defining Functions

Julia offers three syntactic forms for functions:

```julia
# Standard form
function greet(name)
    return "Hello, $name!"
end

# Short form (for one-liners)
square(x) = x^2

# Anonymous (lambda) function
double = x -> 2x
```

```julia
# Functions are first-class objects
map(square, [1, 2, 3, 4])    # [1, 4, 9, 16]
filter(isodd, 1:10)          # [1, 3, 5, 7, 9]
```

```julia
# Apply with do-block syntax
map(1:5) do x
    x^2 + 1
end
```

## 3.2 Arguments and Return Values

```julia
# Positional arguments
function power(base, exponent)
    return base^exponent
end
power(2, 10)    # 1024
```

```julia
# Default values
function power_default(base, exponent=2)
    return base^exponent
end
power_default(5)        # 25
power_default(5, 3)     # 125
```

```julia
# Keyword arguments (after semicolon)
function create_grid(; rows=10, cols=10, fill_value=0.0)
    return fill(fill_value, rows, cols)
end
create_grid(rows=3, cols=4, fill_value=1.0)
```

```julia
# Varargs (variable number of arguments)
function mysum(args...)
    total = 0
    for a in args
        total += a
    end
    return total
end
mysum(1, 2, 3, 4, 5)    # 15
```

```julia
# Multiple return values (as tuple)
function minmax_custom(v)
    return minimum(v), maximum(v)
end
lo, hi = minmax_custom([3, 1, 4, 1, 5, 9])
println("min=$lo, max=$hi")
```

## 3.3 Type Annotations

```julia
# Annotate argument types
function add(x::Number, y::Number)
    return x + y
end

add(3, 4.5)        # 7.5 — works (Int64 <: Number, Float64 <: Number)
# add("a", "b")   # MethodError — String is not a Number
```

```julia
# Parametric type annotations
function first_element(v::Vector{T}) where T
    return v[1]::T
end

first_element([10, 20, 30])
```

## 3.4 Multiple Dispatch

Multiple dispatch is Julia's defining feature. A single function can have many *methods*, selected based on the types of *all* arguments:

```julia
# Define an abstract type and concrete subtypes
abstract type Shape end

struct Circle <: Shape
    radius::Float64
end

struct Rectangle <: Shape
    width::Float64
    height::Float64
end

struct Triangle <: Shape
    base::Float64
    height::Float64
end

# Define area() with multiple methods
area(c::Circle) = π * c.radius^2
area(r::Rectangle) = r.width * r.height
area(t::Triangle) = 0.5 * t.base * t.height

# Julia dispatches based on the argument type
println(area(Circle(5.0)))           # 78.539...
println(area(Rectangle(3.0, 4.0)))   # 12.0
println(area(Triangle(6.0, 3.0)))    # 9.0

# Check all methods of a function
methods(area)
```

## 3.5 Closures

```julia
function make_counter(start=0)
    count = start
    increment() = (count += 1; count)
    reset() = (count = start; count)
    return (increment=increment, reset=reset)
end

c = make_counter(10)
println(c.increment())    # 11
println(c.increment())    # 12
println(c.increment())    # 13
println(c.reset())        # 10
```

```julia
# Closures are common with map/filter
threshold = 0.5
data = rand(10)
filtered = filter(x -> x > threshold, data)

# Adder factory
make_adder(n) = x -> x + n
add5 = make_adder(5)
add5(10)         # 15
```

## 3.6 Function Composition and Piping

```julia
# Composition operator ∘ (\circ + Tab)
f = sqrt ∘ abs
f(-16)            # 4.0
```

```julia
# Pipe operator |>
-16 |> abs |> sqrt    # 4.0

# Chain operations on data
result = [3, 1, 4, 1, 5, 9, 2, 6] |>
    sort |>
    unique |>
    reverse
println(result)
```

```julia
# With anonymous functions in the pipe
[1, 2, 3, 4, 5] |>
    xs -> filter(isodd, xs) |>
    xs -> map(x -> x^2, xs) |>
    sum
```

## 3.7 Metaprogramming Basics

```julia
# An expression is a Julia data structure
ex = :(2 + 3 * x)
typeof(ex)         # Expr
dump(ex)
```

```julia
# Evaluate an expression
x = 10
eval(:(2 + 3 * x))   # 32
```

```julia
# Useful built-in macros
@show 2 + 3        # prints: 2 + 3 = 5, returns 5
@time sum(rand(10^7))  # prints elapsed time and allocations
@assert 1 + 1 == 2 "Math is broken!"
```

```julia
# @elapsed
t = @elapsed begin
    A = rand(1000, 1000)
    B = A * A'
end
println("Matrix multiply took $t seconds")
```

## 3.8 Scope Rules

```julia
# Global scope
x_global = 10

# Local scope (functions)
function demo()
    y = 20        # local to demo
    println("x_global = $x_global")
    return y
end
demo()

# let blocks create a fresh scope
let
    temp = 42
    println("temp = $temp")
end
# temp is not defined here
```
