## 4.3 Indexing and Slicing

```julia
A = [10 20 30; 40 50 60; 70 80 90]

# Single element
A[2, 3]              # 60
A[1]                 # 10 (linear index, column-major)

# Slices
A[1, :]              # first row
A[:, 2]              # second column
A[1:2, 2:3]          # submatrix
```

```julia
# Logical indexing
v = [3, 1, 4, 1, 5, 9, 2, 6]
v[v .> 3]
```

```julia
# Views (no copy — alias to original memory)
A = [10 20 30; 40 50 60; 70 80 90]
v_sub = @view A[1:2, :]
v_sub[1, 1] = 999    # modifies A!
println(A)
```

```julia
# findall, findfirst
A = [10 20 30; 40 50 60; 70 80 90]
findall(x -> x > 50, A)
findfirst(x -> x > 50, A)
```

## 4.4 Broadcasting: The Dot Syntax

```julia
# Element-wise operations with dot
a = [1, 2, 3]
b = [10, 20, 30]

a .+ b               # [11, 22, 33]
a .* b               # [10, 40, 90]
a .^ 2               # [1, 4, 9]
sin.(a)              # [sin(1), sin(2), sin(3)]
```

```julia
# Broadcasting a scalar
a = [1, 2, 3]
a .+ 10

# Fused broadcasting with @.
x = [1.0, 2.0, 3.0]
@. sin(x) + cos(x)^2
```

```julia
# Broadcasting with matrices
A = [1 2; 3 4]
v = [10, 20]

A .+ v               # adds v to each column
A .+ v'              # adds v' to each row
```

## 4.6 Linear Algebra

```julia
using LinearAlgebra

A = [4.0 1.0; 2.0 3.0]
b = [1.0, 2.0]

# Basic operations
A'                      # conjugate transpose
det(A)                  # determinant
tr(A)                   # trace
rank(A)
```

```julia
# Solving linear systems: Ax = b
A = [4.0 1.0; 2.0 3.0]
b = [1.0, 2.0]
x = A \ b               # much faster and more stable than inv(A) * b
println("Solution: $x")
println("Verify: A*x ≈ b → $(A * x ≈ b)")
```

```julia
# Eigenvalues and eigenvectors
A = [4.0 1.0; 2.0 3.0]
vals, vecs = eigen(A)
println("Eigenvalues: $vals")
println("Eigenvectors:\n$vecs")
```

```julia
# SVD
A = [4.0 1.0; 2.0 3.0]
U, S, V = svd(A)
println("Singular values: $S")
```

```julia
# Norms
b = [1.0, 2.0]
norm(b)          # L2 norm
norm(b, 1)       # L1 norm
norm(b, Inf)     # infinity norm

# Dot product and cross product
dot([1, 2, 3], [4, 5, 6])
cross([1, 0, 0], [0, 1, 0])
```

```julia
# Special matrix types
D = Diagonal([1, 2, 3])
A = [4.0 1.0; 1.0 3.0]
S = Symmetric(A)
U = UpperTriangular([4.0 1.0; 0.0 3.0])
```

## 4.7 Matrix Factorizations

```julia
using LinearAlgebra

A = rand(5, 5)
A = A + A'  # make symmetric
A += 5I     # ensure positive definiteness

# LU factorization
F = lu(A)
println("L*U ≈ A[p,:] → $(F.L * F.U ≈ A[F.p, :])")
```

```julia
# Cholesky factorization
C = cholesky(A)
println("L*L' ≈ A → $(C.L * C.L' ≈ A)")
```

```julia
# QR factorization
Q, R = qr(rand(5, 5))
println("Q is $(size(Q)), R is $(size(R))")
```

## 4.8 Sparse Arrays

```julia
using SparseArrays

# Create sparse matrices
S = sparse([1, 2, 3, 1], [1, 2, 3, 3], [10, 20, 30, 5], 3, 3)
println(S)
```

```julia
# Convert dense to sparse
A_dense = [1 0 0; 0 2 0; 0 0 3]
A_sparse = sparse(A_dense)

# Memory comparison
A_big = zeros(1000, 1000)
A_big[1,1] = 1.0; A_big[500,500] = 2.0; A_big[1000,1000] = 3.0
A_sp = sparse(A_big)

println("Dense size: $(Base.summarysize(A_big)) bytes")
println("Sparse size: $(Base.summarysize(A_sp)) bytes")
```
