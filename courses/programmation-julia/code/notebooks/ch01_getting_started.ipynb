## 1.1 Why Julia?

Julia solves the *two-language problem*: scientists prototype in a high-level language (Python, Matlab, R) and then rewrite performance-critical code in C or Fortran. Julia eliminates this step.

> *"We want a language that's open source, with a liberal license. We want the speed of C with the dynamism of Ruby."* — Bezanson, Karpinski, Shah, Edelman (2012)

**Key features:**
1. Just-in-time (JIT) compilation via LLVM
2. Multiple dispatch
3. Rich type system
4. Metaprogramming
5. 10,000+ registered packages

> **Tip:** Julia uses 1-based indexing, like Matlab and R.

```julia
# A simple computation
function fibonacci(n)
    n <= 1 && return n
    a, b = 0, 1
    for _ in 2:n
        a, b = b, a + b
    end
    return b
end

# Print the first 15 Fibonacci numbers
for i in 0:14
    println("F($i) = $(fibonacci(i))")
end
```

## 1.8 Unicode in Julia

Julia supports Unicode variable names and operators. Type LaTeX-like sequences then press Tab:

```julia
# Type \alpha then press Tab to get α
α = 0.05
β = 1 - α

# Type \sqrt then Tab to get √
√2 = √(2)    # equivalent to sqrt(2)

# Greek letters in functions
function Δ(a, b, c)
    return b^2 - 4a*c
end

# Matrix operations with Unicode
A = [1 2; 3 4]
x⃗ = [5, 6]
result = A * x⃗
```
