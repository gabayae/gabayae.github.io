## 8.2 Numerical Integration

```julia
using Pkg
Pkg.add("QuadGK")
```

```julia
using QuadGK

# Integrate sin(x) from 0 to π
result, error = quadgk(sin, 0, π)
println("∫₀^π sin(x)dx = $result (error ≈ $error)")
```

```julia
# More complex integrals
# Gaussian integral
result, _ = quadgk(x -> exp(-x^2), -Inf, Inf)
println("∫ exp(-x²)dx = $result  (should be √π ≈ $(√π))")
```

## 8.4 Ordinary Differential Equations (ODEs)

```julia
using Pkg
Pkg.add(["OrdinaryDiffEq", "Plots"])
```

```julia
using OrdinaryDiffEq, Plots

# Exponential decay: du/dt = -0.5u, u(0) = 1.0
f(u, p, t) = -0.5 * u
u0 = 1.0
tspan = (0.0, 10.0)

prob = ODEProblem(f, u0, tspan)
sol = solve(prob, Tsit5())  # Tsitouras 5th order method

plot(sol, title="Exponential Decay",
     xlabel="t", ylabel="u(t)", label="u(t)", linewidth=2)
```

```julia
# Lotka-Volterra predator-prey system
function lotka_volterra!(du, u, p, t)
    α, β, δ, γ = p
    x, y = u
    du[1] = α*x - β*x*y     # prey
    du[2] = δ*x*y - γ*y     # predator
end

u0 = [1.0, 1.0]       # initial populations
p = [1.5, 1.0, 0.3, 1.0]  # parameters: α, β, δ, γ
tspan = (0.0, 20.0)

prob = ODEProblem(lotka_volterra!, u0, tspan, p)
sol = solve(prob, Tsit5())

plot(sol, title="Lotka-Volterra",
     xlabel="Time", ylabel="Population",
     label=["Prey" "Predator"], linewidth=2)
```

```julia
# Phase portrait
plot(sol, idxs=(1, 2),
     title="Phase Portrait",
     xlabel="Prey", ylabel="Predator",
     label="trajectory", linewidth=2)
```

## 8.5 The Lorenz Attractor

```julia
function lorenz!(du, u, p, t)
    σ, ρ, β = p
    du[1] = σ * (u[2] - u[1])
    du[2] = u[1] * (ρ - u[3]) - u[2]
    du[3] = u[1] * u[2] - β * u[3]
end

u0 = [1.0, 0.0, 0.0]
p = [10.0, 28.0, 8/3]
tspan = (0.0, 100.0)

prob = ODEProblem(lorenz!, u0, tspan, p)
sol = solve(prob, Tsit5())

# 3D plot
plot(sol, idxs=(1, 2, 3),
     title="Lorenz Attractor",
     xlabel="x", ylabel="y", zlabel="z",
     label="", linewidth=0.5)
```

## 8.6 Optimization

```julia
using Pkg
Pkg.add("Optim")
```

```julia
using Optim

# Minimize Rosenbrock function: f(x,y) = (1-x)² + 100(y-x²)²
rosenbrock(x) = (1.0 - x[1])^2 + 100.0 * (x[2] - x[1]^2)^2

result = optimize(rosenbrock, [0.0, 0.0], NelderMead())
println("Minimum at: $(Optim.minimizer(result))")
println("Minimum value: $(Optim.minimum(result))")
```

```julia
# With gradient (faster convergence)
using Optim

function rosenbrock_grad!(G, x)
    G[1] = -2.0 * (1.0 - x[1]) - 400.0 * (x[2] - x[1]^2) * x[1]
    G[2] = 200.0 * (x[2] - x[1]^2)
end

result = optimize(rosenbrock, rosenbrock_grad!, [0.0, 0.0], BFGS())
println("BFGS minimum at: $(Optim.minimizer(result))")
```

```julia
# Constrained optimization with bounds
result = optimize(rosenbrock, [-5.0, -5.0], [5.0, 5.0], [0.0, 0.0], Fminbox(BFGS()))
println("Constrained min at: $(Optim.minimizer(result))")
```
