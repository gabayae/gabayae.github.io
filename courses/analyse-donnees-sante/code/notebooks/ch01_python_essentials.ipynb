## 3. Control flow

### Blood pressure classification

```python
systolic = 148

if systolic >= 140:
    category = "Stage 2 Hypertension"
elif systolic >= 130:
    category = "Stage 1 Hypertension"
elif systolic >= 120:
    category = "Elevated"
else:
    category = "Normal"

print(f"BP {systolic} mmHg -> {category}")
```

### Diabetes screening loop

```python
patients = [
    {"id": "P001", "age": 34, "hba1c": 5.4},
    {"id": "P002", "age": 67, "hba1c": 7.8},
    {"id": "P003", "age": 52, "hba1c": 6.1},
    {"id": "P004", "age": 29, "hba1c": 5.0},
    {"id": "P005", "age": 73, "hba1c": 9.1},
]

for p in patients:
    if p["hba1c"] >= 6.5:
        status = "Diabetic"
    elif p["hba1c"] >= 5.7:
        status = "Pre-diabetic"
    else:
        status = "Normal"
    print(f'  {p["id"]} (age {p["age"]}): HbA1c = {p["hba1c"]} -> {status}')
```

## 4. Functions

### BMI calculator

```python
def bmi(weight_kg, height_m):
    """Calculate Body Mass Index."""
    return weight_kg / (height_m ** 2)

def bmi_category(bmi_value):
    """Classify BMI (WHO categories)."""
    if bmi_value < 18.5: return "Underweight"
    elif bmi_value < 25.0: return "Normal weight"
    elif bmi_value < 30.0: return "Overweight"
    else: return "Obese"

value = bmi(82, 1.75)
print(f"BMI: {value:.1f} ({bmi_category(value)})")
```
