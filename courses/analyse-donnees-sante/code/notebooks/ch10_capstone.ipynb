## 3. Getting started: load and explore your dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid", font_scale=1.1)
```

### Example: Project 1 --- Diabetes risk factor analysis

```python
# Load the Pima Indians Diabetes dataset
url = ("https://raw.githubusercontent.com/jbrownlee/Datasets/"
       "master/pima-indians-diabetes.data.csv")
columns = ["pregnancies", "glucose", "blood_pressure", "skin_thickness",
           "insulin", "bmi", "diabetes_pedigree", "age", "outcome"]
df = pd.read_csv(url, names=columns)
print(f"Shape: {df.shape}")
df.head()
```

```python
# Basic exploration
print(df.info())
print()
df.describe()
```

### Example: Project 2 --- COVID-19 in Africa

```python
# Load Our World in Data COVID-19 dataset
owid_url = ("https://raw.githubusercontent.com/owid/covid-19-data/"
            "master/public/data/owid-covid-data.csv")
covid = pd.read_csv(owid_url)
print(f"Shape: {covid.shape}")
print(f"Countries: {covid['location'].nunique()}")
covid.head()
```

### Example: Project 4 --- Heart disease prediction

```python
# Load UCI Heart Disease dataset
heart_url = ("https://raw.githubusercontent.com/raphaelfontenelle/"
             "heart-disease-uci/main/heart.csv")
heart = pd.read_csv(heart_url)
print(f"Shape: {heart.shape}")
print(f"Heart disease prevalence: {heart['target'].mean():.1%}")
heart.head()
```

## 4. Report template

Your written report should follow this structure (5-8 pages):

1. **Introduction** (0.5-1 page): health context, research question, dataset description.
2. **Data description** (1 page): sources, variables, summary statistics, missing data.
3. **Methods** (1 page): cleaning steps, statistical tests, models, validation strategy.
4. **Results** (1.5-2 pages): figures, tables, test statistics, p-values, CIs.
5. **Discussion** (1-1.5 pages): clinical meaning, comparison with literature, implications.
6. **Limitations** (0.5 page): data quality, generalizability, future work.
