## 1. Loading real health data

```python
import pandas as pd
import numpy as np

# Gapminder: life expectancy, population, GDP per capita
url = "https://raw.githubusercontent.com/datasets/gapminder/main/data/gapminder.csv"
gm = pd.read_csv(url)
print(f"Shape: {gm.shape}")
gm.head()
```

## 2. Exploring the data

```python
print("Columns:", gm.columns.tolist())
print("
Data types:")
print(gm.dtypes)
print(f"
Countries: {gm["country"].nunique()}")
print(f"Continents: {gm["continent"].unique()}")
```

```python
gm.describe()
```

## 3. Filtering rows

```python
# African countries only
africa = gm[gm["continent"] == "Africa"]
print(f"African countries: {africa['country'].nunique()}")

# Most recent year
latest = gm[gm["year"] == gm["year"].max()]
print(f"Latest year: {gm['year'].max()}")
```

```python
# Countries with life expectancy below 50
low_le = latest[latest["lifeExp"] < 50]
print(f"Countries with LE < 50: {len(low_le)}")
low_le[["country", "continent", "lifeExp"]].sort_values("lifeExp")
```

## 4. Sorting and ranking

```python
# Top 10 life expectancy
top10 = latest.sort_values("lifeExp", ascending=False).head(10)
top10[["country", "continent", "lifeExp"]]
```

## 5. Grouping and aggregation

```python
# Life expectancy by continent
latest.groupby("continent")["lifeExp"].agg(
    ["mean", "median", "std", "min", "max", "count"]
).round(1).sort_values("mean", ascending=False)
```

## 6. Creating new columns

```python
# Life expectancy category
gm["le_category"] = pd.cut(
    gm["lifeExp"],
    bins=[0, 50, 65, 75, 100],
    labels=["Very low (<50)", "Low (50-65)", "Medium (65-75)", "High (>75)"]
)

# Cross-tabulation
latest_full = gm[gm["year"] == gm["year"].max()]
pd.crosstab(latest_full["le_category"], latest_full["continent"])
```
